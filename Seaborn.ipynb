```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('Pokemon.csv',index_col=0, encoding='ISO-8859-1')
df.head()
```

```python
sns.lmplot(x='Attack',y='Defense', data=df,
          fit_reg=False,  # No regression line
          hue='Stage')   # Color by evolution stage

#Tweak with Matplotlib
plt.xlim(0,None)
plt.ylim(0,None)
```

```python
# Boxplot
stats_df = df.drop(['Total','Stage','Legendary'], axis=1)
sns.boxplot(data=stats_df)
```

```python
sns.set_style('whitegrid')
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]


sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
sns.swarmplot(x='Type 1',y='Attack', data=df, palette=pkmn_type_colors)
```

```python
plt.figure(figsize=(10,6))

sns.violinplot(x='Type 1', y='Attack', data=df, inner=None, # Remove bars
              palette=pkmn_type_colors)

sns.swarmplot(x='Type 1', y='Attack', data=df, color='k', # Makes points black
              alpha=0.7) # and slightly transparent

plt.title('Attack by Type')
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,6))
sns.regplot(x='Attack',y='Defense', data=df, fit_reg=True, ax=ax1)
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors, ax=ax2)
```
